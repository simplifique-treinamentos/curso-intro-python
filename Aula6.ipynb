```python
caminho = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'
```

```python
df_diabetes = pd.read_csv(caminho, header=None, sep="," , decimal=".")
df_diabetes.head()
```

```python
#Renomeio as colunas
df_diabetes.columns = ['preg','plas','pres','skin','test','mass','pedi','age','class']
df_diabetes.head()
```

```python
#Renomeio as colunas para português
df_diabetes.rename( columns = {'preg': 'num_gravidez',
                               'plas': 'glicose',
                               'pres': 'pressao_dias',
                               'skin': 'espessura_pele',
                               'test': 'teste_insulina',
                               'mass': 'imc',
                               'pedi': 'diabetes_pedigree_function',
                               'age': 'idade',
                               'class': 'diabetes'
                               }, inplace = True
                  )

df_diabetes.head()
```
