# Recategorizao de Variveis - Mtodo cut()

```python
df_diabetes['gravidez_categorica'] = pd.cut(x=df_diabetes['num_gravidez'],
                                              bins=[0, 2, 5, 100],
                                              labels=['0 a 2', '3 a 5', 'mais de 5'],
                                              include_lowest= True
                                            )
```

```python
df_diabetes['gravidez_categorica'].value_counts()
```

# Exerccio - Recategorizar a varivel de desfecho

```python
caminho_cvd = 'https://raw.githubusercontent.com/simplifique-treinamentos/datasets/master/hearts_processed_cleveland.csv'
```

```python
df_cvd = pd.read_csv(caminho_cvd, sep=";", decimal=".")
df_cvd.head()
```

```python
df_cvd.num.value_counts()
```

**Recategorizando a varivel num (desfecho) em doente e no doente**

```python
#@title
df_cvd['doente'] = pd.cut(x=df_cvd['num'],
                                              bins=[0, 0.9, 5], # De 0 a 0.9 - No doente. De 1 a 5 - Doente
                                              labels=['0', '1'],
                                              include_lowest= True
                                            )
df_cvd['doente'].value_counts()
```
